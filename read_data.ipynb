```python
household_income = pd.read_csv("Individual income in Juneau.csv")
household_income.head()
```

```python
household_income = household_income[['Variable observation date', 'Variable observation value', 'Variable properties name']]
household_income.head()
```

```python
household_income_uniqueval = household_income['Variable properties name'].unique()
household_income_uniqueval
```

```python
income_mapping = {
    "Population With an Income of Up to $9,999": 5,
    "Population With an Income Between $10,000 and $14,999": 12.5,
    "Population With an Income Between $15,000 and $24,999": 20,
    "Population With an Income Between $25,000 and $34,999": 30,
    "Population With an Income Between $35,000 and $49,999": 42.5,
    "Population With an Income Between $50,000 and $64,999": 57.5,
    "Population With an Income Between $65,000 and $74,999": 70,
    "Population With an Income of $75,000 or More": 75
}

household_income['Average Income per 1000'] = household_income['Variable properties name'].map(income_mapping)
household_income['Total Income'] = household_income['Variable observation value'] * household_income['Average Income per 1000']
household_income = household_income.groupby('Variable observation date')['Total Income'].sum().reset_index()
plt.plot(household_income['Variable observation date'], household_income['Total Income'], marker='o')
plt.title('Individual Income Over Time')
plt.xlabel('Year')
plt.ylabel('Total Income')
plt.grid(True)
plt.show()
```

```python
number_of_visitor = {
    "Year": [2019, 2020, 2021, 2022, 2023],
    "Number of Visitors": [1330000, 48, 124600, 1200000, 1650000]
}

number_of_visitor = pd.DataFrame(number_of_visitor)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(number_of_visitor["Year"], number_of_visitor["Number of Visitors"], marker='o', linestyle='-', color='b')

plt.title("Number of Visitors to Juneau (2019-2023)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Visitors", fontsize=12)

plt.xticks(number_of_visitor["Year"], fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True)

plt.tight_layout()
plt.show()
```
